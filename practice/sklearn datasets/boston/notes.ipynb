```python
# load a dataset from sklearn
from sklearn import datasets

boston = datasets.load_boston()
```

```python
# numpy -> pandas dataframe
boston_data = boston.data
boston_target = boston.target

# ---

import numpy as np
import pandas as pd

boston_data = pd.DataFrame(boston_data, columns=boston.feature_names)
boston_target = pd.DataFrame(boston_target, columns=["MEDV"])
```

```python
# create a test set
from sklearn.model_selection import train_test_split

boston_X_train, boston_X_test, boston_y_train, boston_y_test = train_test_split(
    boston_data, boston_target, test_size=0.2, random_state=42)
```

```python
# plot some figures
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
%matplotlib inline

scatter_matrix(boston_X_train, figsize=(12, 8))
```

```python
# standard scaler
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
boston_X_train_prepared = std.fit_transform(boston_X_train)
```

```python
# pipeline
from sklearn.pipeline import Pipeline
```

```python
# cv

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(boston_X_train_prepared, boston_y_train)

# ---

from sklearn.model_selection import cross_val_score

mse = cross_val_score(lin_reg, boston_X_train_prepared, boston_y_train, cv=3, scoring="neg_mean_squared_error")
```
